```python
# 1) Elabore um código que imprima a mensagem *"Olá terráqueos!!"* na tela.

print("Olá terráqueos")
```

```python
# 2) Elabore um código que peça um número e então mostre a mensagem: "O número informado foi [número]."

i = int(input("Digite um número "))

print("O número informado foi ",i)
```

```python
# 3) Elabore um código que peça dois números e imprima a soma.

i = int(input("Digite o Primeiro número: "))
j = int(input("Digite o segundo número: "))

print("A soma dos dois número foi equivalente a: ", i + j)
```

```python
# 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.

nota_1 = int(input ("Insira a Primeira Nota: "))
nota_2 = int(input ("Insira a Segunda Nota: "))
nota_3 = int(input ("Insira a Terceira Nota: "))
nota_4 = int(input ("Insira a Quarta Nota: "))

media = (nota_1 + nota_2 + nota_3 + nota_4) / 4

print("A média equivalente final das notas foi: ", media)
```

```python
# 5) Elabore um código que converta metros para centímetros.

metros = float(input ("Insira um valor em metros: "))

conversao = metros * 100

print("O valor convertido foi: ", conversao, "centímetros")
```

```python
# 6) Elabore um código que peça o raio de um círculo, calcule e mostre sua área.

import math

raio = float(input ("Digite um valor para o raio de um círculo: "))

area_circulo = (raio)**2 * math.pi

print("A área do círculo é: ", area_circulo)

```

```python
# 7) Elabore um código que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

lado_quadrado = float(input ("Insira o lado do quadrado: "))

area_quadrado = lado_quadrado**2

print("A dobro da área do quadrado é: ", area_quadrado**2)
```

```python
# 8) Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.

ganho_por_hora = float(input("Informe quanto voce recebe por hora: "))
numero_horas_total_mes = float(input("Agora informe o número de horas trabalhadas no mês: "))

salario_total = ganho_por_hora * numero_horas_total_mes


print("O total do seu salário no referido mês é ", salario_total)
```

```python
# 9) Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. ** Onde C = 5 * ((F-32) / 9)**.

F = float(input ("Insira um valor de temperatura em graus Fahrenheit: "))

conversao_em_C = 5 * ((F-32) / 9)

print("A temperatura em graus Celsius é equivalente a ", conversao_em_C)

```

```python
# 10) Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

C = float(input ("Insira um valor de temperatura em graus Celsius: "))

conversao_em_F = (C * 9/5) + 32

print("A temperatura em graus Fahrenheit é equivalente a ", conversao_em_F)
```

```python
# 11) Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:

numero_1 = int(input("Insira o Primeiro número Inteiro: "))
numero_2 = int(input("Insira o Segundo número Inteiro: "))
numero_3 = float(input("Insira um número Real: "))

# a) o produto do dobro do primeiro com metade do segundo.

calculo_a = (numero_1 * 2) * (numero_2 / 2)

print("O produto do dobro do primeiro com metade do segundo é: ", calculo_a)

# b) a soma do triplo do primeiro com o terceiro.

calculo_b = (numero_1 * 3) + numero_3

print("A soma do triplo do primeiro com o terceiro é: ", calculo_b)

# c) o terceiro elevado ao cubo.

calculo_c = numero_3**3

print("O terceiro elevado ao cubo é: ", calculo_c)

```

```python
# 12) Elabore um código que receba os dados de entrada a altura de uma pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula: (72.7*altura) - 58

altura = float(input("Informe a sua Altura: "))

peso_ideal = (72.7 * altura) - 58

print("O peso ideal é: ", peso_ideal)

```
